```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/sample_data/Electric_Vehicle_Population_Data.csv")
df.head()
```

3. Data Cleaning


```python
from ctypes import string_at
missing_values = df.isnull().sum()

df = df.dropna()

duplicates = df.duplicated().sum()

duplicates = df.duplicated().sum()

df['Electric Range'] = df['Electric Range'].astype(int)
df['Model Year'] = df['Model Year'].astype(int)

df.head()
```

```python
plt.plot(df['Make'].unique(),df['Make'].value_counts())
plt.xticks(rotation=90)
```

4.  Filtering by State

5.EVs provide an Alternative to Fuel or not?

```python
%matplotlib inline

cafv_eligibility_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()

plt.plot(figsize=(10,6))
plt.plot(cafv_eligibility_counts.index ,cafv_eligibility_counts.values)
plt.title('CAFV Eligibility Distribution')
plt.ylabel('Number of Vehicles')
plt.xlabel('CAFV Eligibility')
plt.xticks(rotation=90)
plt.show()

```

Visualizing Electric Utility

```python
%matplotlib inline

top_10_utilities = df['Electric Utility'].value_counts().head(10)
plt.figure(figsize=(12,6))
plt.bar(top_10_utilities.index, top_10_utilities.values)
plt.title('Top 10 Electric Utilities Serving EVs')
plt.ylabel('Number of Vehicles')
plt.xlabel('Electric Utility')
plt.xticks(rotation=90)
plt.show()
```
